```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Data meanings

- Country = País
- League = Liga (Brasileira)
- Season = Temporada
- Data = Data do Jogo
- Time = Horário do Jogo
- Home = Time da Casa
- Away = Time visitante
- HG = Gols time da casa
- AG = Gols time visitante
- Res = Resultado do jogo (D = Draw, H = Home Win, A = Away win)
- PH = Probablidade de vitória do time da casa provida pela Pinacle
- PD: Probablidade de empate provida pela Pinacle
- MaxH = Probablidade de vitória do time da casa provida pela OddsPortal 
- MaxD = Probablidade de empate provida pela OddsPortal
- MaxA = Probablidade vitória time visitante provida pela OddsPortal
- AvgH = Probablidade média de vitória em casa provida pela OddsPortal
- AvgD = Probablidade média de empate provida pela OddsPortal
- AvgA = Probablidade média de vitória pelo time visitante provida pela OddsPortal
### Pinacle e Odds Portal são, ambas, casas de apostas inglesas.



```python
data = pd.read_csv("data/BRA.CSV")
data
```

###  Avarege, median and std of HG and AG


```python
media_HG = data.HG.mean() 
median_HG = data.HG.median()
std_HG = data.HG.std()

print("The home team´s avarege number of goals:", media_HG)
print("The home team´s median number of goals:", median_HG)
print("The home team´s std number of goals:", std_HG)
```

```python
media_AG = data.AG.mean() 
median_AG = data.AG.median()
std_AG = data.AG.std()

print("The away team´s avarege number of goals:", media_AG)
print("The away team´s median number of goals:", median_AG)
print("The away team´s std number of goals:", std_AG)
```

### Graph bar of results distribuation (D, H, A)

```python
results = data.Res.value_counts()
index_results = results.index

plt.style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(12,5))
graph = ax.bar(index_results, results, color = ('green', 'blue', 'red'))    
ax.set_title("Results distribuation graph in the Brazilian football league\n2012-2022")
ax.set_xlabel("Result type (H-HomeWin, D-Draw, A-AwayWin)", fontsize=12)
ax.set_ylabel("Number of results", fontsize=12)
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)
ax.grid(linestyle='--')
ax.bar_label(graph, fmt="%.01f", size=10, label_type='edge')

plt.show()
```

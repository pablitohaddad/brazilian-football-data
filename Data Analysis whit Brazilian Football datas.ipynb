```python
data = pd.read_csv("data/BRA.CSV")
data
```

### 1. Avarege, median and std of HG and AG


```python
media_HG = data.HG.mean() # Media HG
median_HG = data.HG.median() # Median HG
std_HG = data.HG.std() # Std HG

print("The home team´s avarege number of goals:", media_HG)
print("The home team´s median number of goals:", median_HG)
print("The home team´s std number of goals:", std_HG)
```

```python
media_AG = data.AG.mean() # Media AG
median_AG = data.AG.median() # Median AG
std_AG = data.AG.std() # Std AG

print("The away team´s avarege number of goals:", media_AG)
print("The away team´s median number of goals:", median_AG)
print("The away team´s std number of goals:", std_AG)
```

### 2. Graph bar of results distribuation (D, H, A)

```python
results = data.Res.value_counts() # Values counts of D, H, A
index_results = results.index # Index (D, H, A)

plt.style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(12,5))
graph = ax.bar(index_results, results, color = ('green', 'blue', 'red')) # X = index_results, Y = results
ax.set_title("Results distribuation graph in the Brazilian football league\n2012-2022")
ax.set_xlabel("Result type (H-HomeWin, D-Draw, A-AwayWin)", fontsize=12)
ax.set_ylabel("Number of results", fontsize=15)
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=10)
ax.grid(linestyle='--')
ax.bar_label(graph, fmt="%d wins", size=10, label_type='edge')

plt.show()
```

### 3. The correlation between the probability of winning at home and the average number of goals scored at home

```python
# Correlation whit Pearson´s Coefficient 

correlation = data['AvgH'].corr(data['HG'])
correlation
```

```python
# Outliers AvgH

data.boxplot(['AvgH'])
```

```python
# Outliers HG

data.boxplot(['HG'])
```

```python
# Remove Outliers AvgH
valor = data['AvgH']
Q1 = valor.quantile(.25)   
Q3 = valor.quantile(.75)  
IIQ = Q3 - Q1 
limite_inferior = Q1 - 1.5 * IIQ 
limite_superior = Q3 + 1.5 * IIQ

selecao = (valor >= limite_inferior) & (valor <= limite_superior) 
data_outlier_avgh= data[selecao]

data_outlier_avgh.boxplot(['AvgH'])
```

```python
# With Outlier
plt.style.use('fivethirtyeight')

fig, ax = plt.subplots(figsize=(12,5))
ax.scatter(data.HG, data.AvgH) 
ax.set_title("Correlation study between probability of victory and goals scored at home")
ax.set_xlabel("Average home team win Probability ", fontsize=12)
ax.set_ylabel("Avarege home team goals", fontsize=15)
ax.xaxis.set_tick_params(labelsize=15)
ax.yaxis.set_tick_params(labelsize=10)
ax.grid(linestyle='--')
```
